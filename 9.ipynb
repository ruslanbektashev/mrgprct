##Результаты тестирования
В университете провели тестирование студентов по программированию с помощью Яндекс.Контеста, и теперь преподавателям хочется узнать, какие языки программирования использовали студенты в ходе тестирования. Из Яндекс.Контеста можно получить текстовый файл с описанием посылок. Для каждого языка программирования вычислите количество успешных посылок на этом языке.
#####Формат ввода:
До $1000$ строк такого вида: ФИО студента \t название задачи \t время посылки \t язык программирования \t результат
Здесь \t — символ табуляции.
#####Формат вывода:
Для каждого языка программирования с ненулевым числом успешных посылок выведите строку вида 'название количество' (через пробел). Строки упорядочите по убыванию счёта, языки с одинаковым числом успешных посылок упорядочите лексикографически.
###Пример
#####Ввод:
---
    Augustus Marcus Maximus B	17.09.2015 06:31 AM	Pascal	PE
    Julius Justus Hortator	D	13.09.2015 01:07 PM	C++	WA
    Marcus Augustus Maximus	C	18.09.2015 08:51 PM	Java	WA
    Urban Augustus Maximus	D	18.09.2015 11:21 PM	C++	PE
    Magnus Cyprian Albus	B	16.09.2015 01:05 AM	C#	WA
    Magnus Octavius Albus	F	10.09.2015 11:01 PM	Pascal	OK
---
#####Вывод:
---
    Pascal 1
---

```python
results = []
successors = {}

while True:
    try:
        columns = raw_input().split('\t')
    except EOFError:
        break
    if columns:
        if len(columns) == 5:
            results.append({'name': columns[0],
                            'task': columns[1],
                            'date': columns[2],
                            'lang': columns[3],
                            'stat': columns[4]})
        else:
            break
    else:
        break
        
for result in results:
    if result['stat'] == "OK":
        if result['lang'] in successors:
            successors[result['lang']] += 1
        else:
            successors[result['lang']] = 1

for lang, count in successors.items():
    print lang, count
```

##Оценки
Преподаватель Васиной группы наконец-то начал объявлять оценки за прошедшие работы. Так как важен не набор оценок, а средняя, то после каждого объявления студент, которому сообщили новую оценку, хочет знать среднее своих оценок по уже объявленным работам. Напишите программу, которая избавит Васиных одногруппников от лишнего ручного счета.
#####Формат ввода:
Первая строка входного файла - целое число $N$ от $0$ до $10^5$ - общеее количество оценок.
Далее идут $N$ строк, каждая из которых содержит фамилию очередного студента (строка из латинских букв длиной от $1$ до $20$ символов) и его оценку - целое число от $0$ до $10^9$.
#####Формат вывода:
Выведите $N$ строк. $k$-я строка должна содержать среднюю оценку студента, которому была выставлена $k$-я оценка в исходном списке, после объявления $k$ оценок. Средняя оценка округляется до ближайшего целого вниз (то есть, от нее отбрасывается дробная часть).
###Пример
#####Ввод:
---
    5
    petrov 3
    sidorov 8
    petrov 5
    petrov 3
    ivanov 9
---
#####Вывод:
---
    3
    8
    4
    3
    9
---

```python
N = int(raw_input())

marks = {}
output = []

while N > 0:
    N -= 1
    string = raw_input().split()
    if string[0] in marks:
        marks[string[0]].append(int(string[1]))
    else:
        marks[string[0]] = [int(string[1])]
    output.append(str(sum(marks[string[0]]) / len(marks[string[0]])))
    
print "\n".join(output)
```

##Сессия
Пришло время Васиной группе сдавать сессию.
В программе есть $N$ предметов ($0 \leq N \leq 10^4$), по каждому из которому установлен минимальный балл.
Каждый из студентов Васиной группы может пытаться сдать экзамен несколько раз (в том числе и пересдать на более высокую оценку). При этом некоторые одногруппники сдают еще и дополнительные экзамены, которых нет в программе.
Вам нужно для каждого студента, который пытался сдать хотя бы один экзамен, найти число экзаменов, которые он сдал хотя бы на минимальный балл.
#####Формат ввода:
Первая строка содержит число $N$.
Далее идут $N$ строк, каждая из которых содержит название очередного предмета (строку из не более чем $10$ латинских символов) и целое число от $1$ до $10^9$ - минимальный балл, необходимый для успешной сдачи экзамена.
Далее идет целое число $M$, $0 \leq M \leq 5*10^4$  - число попыток сдачи экзаменов.
Далее идут $M$ строк, каждая их которых содержит фамилию очередного студента (строку из не более чем $10$ латинских символов), предмет и балл, полученный за предмет.
#####Формат вывода:
Для каждого студента, пытавшегося сдать хотя бы один экзамен, выведите его фамилию и число успешно сданных им экзаменов.
Список студентов отсортируйте по числу успешных сдач, а при равенстве - по алфавиту.
###Пример
#####Ввод:
---
    3
    algebra 3
    mathan 3
    programming 4
    6
    Ivanov algebra 3
    Sidorov algebra 2
    Ivanov mathan 3
    Ivanov algebra 5
    Petrov programming 4
    Sidorov algebra 4
---
#####Вывод:
---
    Ivanov 2
    Petrov 1
    Sidorov 1
---
